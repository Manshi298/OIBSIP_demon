```python
import seaborn as sns
sns.heatmap(dt.corr(),annot=True)
plt.show()
```

Analyzing data by histogram

```python
fg = px.histogram(dt,x=' Estimated Employed',color='Region',title='Indian Unemploymeny by Histogram',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

Analyzing data by histogram

```python
fg = px.histogram(dt,x=' Estimated Unemployment Rate (%)',color='Region',title='Indian Unemploymeny by Histogram',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

Analyzing Data By Bar Graphs

```python
fg = px.bar(dt,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',
            title='Unemploymeny Rate (State Wise) by Bar Graph',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

```python
fg = px.bar(dt,x=' Date',y=' Estimated Unemployment Rate (%)',color='Region',
            title='Unemploymeny Rate by Bar Graph',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

Analyzing Data By Box Plot

```python
fg = px.box(dt,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',
            title='Unemploymeny Rate (State wise) by Box Plot',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

Analyzing Data By Scatter Plot

```python
fg = px.scatter(dt,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',
                title='Unemploymeny Rate (Statewise) by Scatter Plot',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```
